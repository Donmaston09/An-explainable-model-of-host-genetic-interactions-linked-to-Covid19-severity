```python
from matplotlib import pyplot as plt
import numpy as np
pylab.rcParams['figure.figsize'] = (15.0, 10.0)
plt.rc('font', size=28)
fig = plt.figure()
ax = fig.add_axes([0,0,1,1])
ax.axis('equal')
group = ['asymptomatic', 'severe']
value = [30,70]
ax.pie(value, colors = ['b', 'r'], labels = group,autopct='%1.1f%%')
plt.title('Respiratory or thoracic disease')
plt.savefig(path+'Respiratory_disease_pie_chart.pdf', format="pdf", bbox_inches="tight")
plt.show()
```
